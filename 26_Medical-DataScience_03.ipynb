## リスト p.105

変数は1変数あたり一つの数字を保存したが、リストは一つの変数に複数の数字を保存できる。
```
sample = [ 10.5, 3.8, 5.5, 6.7]
```
これはsampleというリストに、4つの数字を保存する命令文である。
```
print(sample)
```
とすると、sampleというリストに格納された4つ数字がまとめて出力される。

画像を扱う場合は、一つ一つのピクセルのピクセル値をリストに格納することになる。

````python
sample = [ 10.5, 3.8, 5.5, 6.7]
````

````python
print(sample)
````

### リスト要素の参照

リストに格納されたデータから、目的の一つ読み出すには、リストの要素につけられた番号を用いる。
```
print(sample[0])
```
番号は"0番"から始まっていることに注意

プログラミングでは0番目から数え始めているか、1番目から数え始めているかに注意が必要。(統計言語のRや数式処理ソフトMatlabは1からスタート、PythonやC言語などプログラミング言語は0からスタートが多い)

bin幅、bin境界を追加標示する

````python
data = iris_df['petal length (cm)']
counts, bin_edges = np.histogram(data, bins=10)  # デフォルトbins=10

fig, ax = plt.subplots()
ax.hist(data, bins=bin_edges, edgecolor='white', linewidth=0.5)

# bin widthを計算（全bin同一幅なので最初の差分を使用）
bin_width = bin_edges[1] - bin_edges[0]

# bin edgesとbin widthをテキストで表示
edges_str = ', '.join([f'{e:.3f}' for e in bin_edges])
ax.set_title('petal length (cm)')
ax.set_xlabel(f'Bin width: {bin_width:.3f}\nBin edges: [{edges_str}]', fontsize=8)

ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
````

#### ビン境界について

pd.DataFrame.plot()の標準ははbin数10固定で、bin幅が自動設定される。bin edgeが人の常識と異なる数字になるため。読みやすいように0.5の倍数に丸める。

````python
# 最小値と最大値を取得
x_min_data = iris_df['sepal width (cm)'].min()
x_max_data = iris_df['sepal width (cm)'].max()

data = iris_df['sepal width (cm)']

# bin幅0.5に丸めたedgeを生成
start_boundary = np.floor(data.min() * 2) / 2
end_boundary   = np.ceil(data.max() * 2) / 2
bin_edges = np.arange(start_boundary, end_boundary + 0.5, 0.5)
bin_width = bin_edges[1] - bin_edges[0]

fig, ax = plt.subplots()
ax.hist(data, bins=bin_edges, edgecolor='white', linewidth=0.5)

edges_str = ', '.join([f'{e:.1f}' for e in bin_edges])
ax.set_title('sepal width (cm)')
ax.set_xlabel(f'Bin width: {bin_width:.1f}\nBin edges: [{edges_str}]', fontsize=8)

ax.spines[['top', 'right']].set_visible(False)
plt.tight_layout()
plt.show()
````

## Pair plotを用いた可視化

### sns.pairplot()によるPair Plot
4つのパラメータを2つずつ選択し、それぞれ種類がわかるようにplotする。

参考: scikit-learnによる機械学習プログラミング入門 - 生命情報実験（情報系課題） https://keio-bioinformatics.github.io/jikken/docs/3-1-iris/

````python
data_multi = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
data_multi['species'] = [iris_data.target_names[i] for i in iris_data.target]
data_multi.sample(10)
sns.pairplot(data_multi, hue='species', markers=["o", "s", "X"],corner=True) #seaborn: `scatterplot` の `markers` に指定できる文字 https://qiita.com/noobar/items/be9008f2eceabeaf4a9a
````
